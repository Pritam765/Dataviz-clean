```python
top_host = room.host_id.value_counts().head(10)
top_host

# which top 5 host_IDs have the most listings on dataset 
```

```python
sns.set(rc={'figure.figsize':(12,6)})
sns.set_style('white')
top_host_df=pd.DataFrame(top_host)
top_host_df.reset_index(inplace=True)
top_host_df.rename(columns={'index':'Host_ID', 'host_id':'P_Count'}, inplace=True)
top_host_df
```

```python
plt=sns.barplot(x="Host_ID", y="P_Count", data=top_host_df,palette='viridis')
plt.set_title('Hosts with the most listings',color='green')
plt.set_ylabel('Count of listings')
plt.set_xlabel('Host IDs')
plt.set_xticklabels(plt.get_xticklabels(), rotation=45)
```

```python
top_reviewed_listings=room.nlargest(10,'number_of_reviews')
top_reviewed_listings

# let's grab 10 most reviewed listings in airbnb
```

```python
price_avrg=top_reviewed_listings.price.mean()
print('Average price per night: {}'.format(price_avrg))
```

```python
# Minimum_nights based room price and room type
import matplotlib.pyplot as plt
import seaborn as sns
sns.barplot(x = 'minimum_nights', y = 'price', hue = 'room_type', data = room)
plt.title("Area based top luxury room")
plt.show()
```

```python
sns.barplot(x = 'neighbourhood_group', y = 'price', hue = 'room_type', data = room)
plt.title("Area based top luxury room")
plt.show()
```
